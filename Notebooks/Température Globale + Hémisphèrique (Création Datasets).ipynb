```python
import pandas as pd
```

```python
# importation du jeu de données des températures de l'hémisphère Nord: northern-hemisphere-TAVG-Trend.txt
headers = ['Year', 'Month', 'Anomaly-NH', 'Uncertainty-NH']

df_North = pd.read_csv('northern-hemisphere-TAVG-Trend.txt', header = 69, sep='\s+', names = headers, index_col=False)
df_North
```

```python
# importation du jeu de données des températures de l'hémisphère Sund: southern-hemisphere-TAVG-Trend.txt
headers_2 = ['Year', 'Month', 'Anomaly-SH', 'Uncertainty-SH']

df_South = pd.read_csv('southern-hemisphere-TAVG-Trend.txt', header = 69, sep='\s+', names = headers_2, index_col=False)
df_South
```

```python
# Fusion des 2 dataframes de températures hémisphériques
df_Hemisphere = df_North.merge(right = df_South, on = ['Year', 'Month'], how = 'outer')
```

```python
df_Hemisphere
```

```python
# Exportation Dataframe de températures_hémisphériques
df_Hemisphere.to_csv("températures_hémisphériques.csv")
```

```python
# importation du jeu de données des températures globales: Complete_TAVG_complete.txt

headers = ['Year', 'Month', 'Anomaly-Gl', 'Uncertainty-Gl']

df_Glob = pd.read_csv('Complete_TAVG_complete.txt', header = 69, sep='\s+', names = headers, index_col=False)
df_Glob
```

```python
# Exportation du Dataframe de températures_globales
df_Glob.to_csv("températures_globales.csv")
```

```python
# Fusion du dataframe Températures globales avec Températures hémisphériques
df_Temp_Commplete = df_Glob.merge(right = df_Hemisphere, on = ['Year', 'Month'], how = 'outer')
```

```python
df_Temp_Commplete
```

```python
# Exportation du Dataframe de températures_globales_hémisphériques
df_Temp_Commplete.to_csv("températures_globales_hémisphériques.csv")
```
