```python
# Importation des packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
# CrÃ©ation du Dataframe

df = pd.read_table('Full_ensemble_median_and_95pct_range_edit.txt')
df
```

```python
# CrÃ©ation d'un graphe avec toutes les donnÃ©es disponibles

fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(111)

s1 = df['Cowtan & Way instrumental target']
s2 = df['Full ensemble median']
s3 = df['Full ensemble 2.5th percentile']
s4 = df['Full ensemble 97.5th percentile']
s5 = df['Cowtan & Way instrumental target 31-year filtered']
s6 = df['31-year filtered full ensemble median']
s7 = df['31-year filtered full ensemble 2.5th percentile']
s8 = df['31-year filtered full ensemble 97.5th percentile']


ax1.plot(s1,color='#33CCFF',label='Cowtan & Way instrumental target')
ax1.plot(s2,color='#FFCC99',label='Full ensemble median')
ax1.plot(s3,color='#FF33CC',label='Full ensemble 2.5th percentile')
ax1.plot(s4,color='#AF33CC',label='Full ensemble 97.5th percentile')
ax1.plot(s5,color='#00bbf9',label='Cowtan & Way instrumental target 31-year filtered')
ax1.plot(s6,color='#f48c06',label='31-year filtered full ensemble median')
ax1.plot(s7,color='#fee440',label='31-year filtered full ensemble 2.5th percentile')
ax1.plot(s8,color='#e85d04',label='31-year filtered full ensemble 97.5th percentile')


#plt.xticks([0,20,40,60,80,100,120,140], ['1880', '1900', '1920', '1940', '1960', '1980', '2000', '2020']);

ax1.legend(loc='upper left');


plt.xlabel('AnnÃ©e');
plt.ylabel('Anomalie');
plt.title("Evolution des anomalies de tempÃ©ratureÂ mensuelle en fonction de l'annÃ©e");
```

```python
# CrÃ©ation d'un graphe avec seulement le Full Ensemble Median

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
s = df['Full ensemble median']
ax.plot(s,label='Full ensemble median');
ax.legend(loc='upper left');
plt.xlabel('AnnÃ©e');
plt.ylabel('Anomalie');
plt.title("Evolution des anomalies de tempÃ©ratureÂ annuelle depuis 2000 ans");
```

```python
# CrÃ©ation d'un graphe avec le Full Ensemble Median et la moyenne filtrÃ©e sur 31 ans.

fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(111)

s1 = df['Full ensemble median']
s2 = df['31-year filtered full ensemble median']

ax1.plot(s1,color='blue',label='Full ensemble median', alpha=0.4)
ax1.plot(s2,color='red',label='31-year filtered full ensemble median')


#plt.xticks([0,20,40,60,80,100,120,140], ['1880', '1900', '1920', '1940', '1960', '1980', '2000', '2020']);

ax1.legend(loc='upper left');


plt.xlabel('AnnÃ©e');
plt.ylabel('Anomalie');
plt.title("Evolution des anomalies de tempÃ©ratureÂ mensuelle en fonction de l'annÃ©e");
```
